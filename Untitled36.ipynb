```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select, WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

# Setup Selenium WebDriver (use Chrome or Firefox)
driver = webdriver.Chrome()  # Make sure chromedriver is installed and in PATH
driver.get("https://www.doccafe.com/jobs/search")

wait = WebDriverWait(driver, 10)

# Wait for Occupation dropdown to load
occupation_dropdown = wait.until(
    EC.presence_of_element_located((By.ID, "occupation_ids"))
)

# Get all Occupation options
occupation_options = occupation_dropdown.find_elements(By.TAG_NAME, "option")

results = []

for option in occupation_options:
    occupation_value = option.get_attribute("value")
    occupation_name = option.text.strip()
    
    if occupation_value == "":
        continue  # Skip empty/default options

    # Select Occupation
    select_occupation = Select(occupation_dropdown)
    select_occupation.select_by_value(occupation_value)
    time.sleep(2)  # Wait for Specialty dropdown to reload

    # Get Specialty dropdown
    specialty_dropdown = wait.until(
        EC.presence_of_element_located((By.ID, "specialty_ids"))
    )
    specialty_options = specialty_dropdown.find_elements(By.TAG_NAME, "option")

    for spec in specialty_options:
        specialty_value = spec.get_attribute("value")
        specialty_name = spec.text.strip()
        
        if specialty_value == "":
            continue

        results.append({
            "Occupation": occupation_name,
            "Specialty": specialty_name,
            "Specialty_ID": specialty_value,
            "Occupation_ID": occupation_value
        })

# Close browser
driver.quit()

# Save to CSV
df = pd.DataFrame(results)
df.to_csv("occupation_specialties.csv", index=False)

print("Scraping completed and saved to 'occupation_specialties.csv'")
```

```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# ✅ Setup ChromeDriver
service = Service('C:\\Users\\chromedriver-win64\\chromedriver.exe')  # Update if needed
driver = webdriver.Chrome(service=service)
wait = WebDriverWait(driver, 20)

# ✅ Target URL
driver.get("https://www.doccafe.com/jobs")
driver.maximize_window()
time.sleep(3)

# ✅ Define Occupations to extract from
occupations = ["CRNA", "Midwife", "Nurse Practitioner", "Physician", "Physician Assistant"]

# ✅ Storage for results
results = []

# ✅ Loop through each Occupation
for occ in occupations:
    try:
        # Click the checkbox for the occupation
        checkbox_xpath = f"//label[.//span[text()='{occ}']]"
        checkbox = wait.until(EC.element_to_be_clickable((By.XPATH, checkbox_xpath)))
        driver.execute_script("arguments[0].scrollIntoView();", checkbox)
        checkbox.click()
        print(f"✅ Selected Occupation: {occ}")
        time.sleep(2)

        # Click the Specialty dropdown
        dropdown = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[contains(text(),'Specialty')]")))
        dropdown.click()
        time.sleep(1)

        # Scroll dropdown listbox to load all options
        listbox = wait.until(EC.presence_of_element_located((By.XPATH, "//div[@role='listbox']")))
        for _ in range(20):
            driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", listbox)
            time.sleep(0.1)

        # Get all options
        options = listbox.find_elements(By.XPATH, ".//div[@role='option']")
        print(f"   ➤ Found {len(options)} specialties under {occ}")

        for opt in options:
            text = opt.text.strip()
            if text and text.lower() != "select specialty":
                results.append((occ, text))

        # Click outside to close dropdown
        driver.find_element(By.TAG_NAME, 'body').click()
        time.sleep(1)

        # Uncheck the current occupation before continuing
        checkbox = wait.until(EC.element_to_be_clickable((By.XPATH, checkbox_xpath)))
        checkbox.click()
        time.sleep(1)

    except Exception as e:
        print(f"❌ Error with occupation {occ}:", e)
        continue

# ✅ Save to CSV
with open("occupation_specialties.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Occupation", "Specialty"])
    writer.writerows(results)

print("\n✅ Saved all specialties per occupation to 'occupation_specialties.csv'")
driver.quit()
```

```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# ✅ Path to your ChromeDriver
service = Service('C:\\Users\\chromedriver-win64\\chromedriver.exe')
driver = webdriver.Chrome(service=service)
wait = WebDriverWait(driver, 20)

# ✅ Occupation URLs
occupation_urls = {
    "CRNA": "https://www.doccafe.com/crna-jobs",
    "Midwife": "https://www.doccafe.com/midwife-jobs",
    "Nurse Practitioner": "https://www.doccafe.com/nurse-practitioner-jobs",
    "Physician": "https://www.doccafe.com/physician-jobs",
    "Physician Assistant": "https://www.doccafe.com/physician-assistant-jobs"
}

results = []

for occupation, url in occupation_urls.items():
    try:
        driver.get(url)
        driver.maximize_window()
        print(f"\n🔎 Processing Occupation: {occupation}")
        time.sleep(3)

        # Click Specialty dropdown
        dropdown = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[contains(text(),'Specialty')]")))
        dropdown.click()
        time.sleep(2)

        # Scroll dropdown list
        listbox = wait.until(EC.presence_of_element_located((By.XPATH, "//div[@role='listbox']")))
        for _ in range(20):
            driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", listbox)
            time.sleep(0.1)

        # Get all specialty options
        options = listbox.find_elements(By.XPATH, ".//div[@role='option']")
        print(f"  🔍 Found {len(options)} specialties under {occupation}")
        for opt in options:
            text = opt.text.strip()
            if text and text.lower() != "select specialty":
                results.append((occupation, text))

        driver.find_element(By.TAG_NAME, 'body').click()
        time.sleep(1)

    except Exception as e:
        print(f"❌ Error processing {occupation}: {e}")

# ✅ Save results to CSV
with open("occupation_specialties.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Occupation", "Specialty"])
    writer.writerows(results)

print("\n✅ Saved all specialties to 'occupation_specialties.csv'")
driver.quit()
```

```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# ✅ Setup ChromeDriver
service = Service('C:\\Users\\chromedriver-win64\\chromedriver.exe')  # Adjust if needed
driver = webdriver.Chrome(service=service)
wait = WebDriverWait(driver, 30)

# ✅ Occupation-specific URLs
occupation_urls = {
    "CRNA": "https://www.doccafe.com/crna-jobs",
    "Midwife": "https://www.doccafe.com/midwife-jobs",
    "Nurse Practitioner": "https://www.doccafe.com/nurse-practitioner-jobs",
    "Physician": "https://www.doccafe.com/physician-jobs",
    "Physician Assistant": "https://www.doccafe.com/physician-assistant-jobs"
}

results = []

for occupation, url in occupation_urls.items():
    try:
        print(f"\n🔄 Processing: {occupation}")
        driver.get(url)
        driver.maximize_window()
        time.sleep(5)  # let the page fully load

        # ✅ Click Specialty dropdown
        dropdown = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[contains(text(),'Specialty')]")))
        dropdown.click()
        time.sleep(2)

        # ✅ Locate the dropdown list container
        listbox = wait.until(EC.presence_of_element_located((By.XPATH, "//div[@role='listbox']")))

        # ✅ Scroll to load all options
        for _ in range(30):
            driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", listbox)
            time.sleep(0.1)

        # ✅ Extract visible options
        options = listbox.find_elements(By.XPATH, ".//div[@role='option']")
        print(f"✅ Found {len(options)} specialties under {occupation}")

        for opt in options:
            text = driver.execute_script("return arguments[0].innerText", opt).strip()
            if text and text.lower() != "select specialty":
                results.append((occupation, text))

        # ✅ Close dropdown
        driver.find_element(By.TAG_NAME, 'body').click()
        time.sleep(1)

    except Exception as e:
        print(f"❌ Error with {occupation}: {e}")

# ✅ Save to CSV
if results:
    with open("occupation_specialties.csv", "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["Occupation", "Specialty"])
        writer.writerows(results)
    print("\n✅ All specialties saved to 'occupation_specialties.csv'")
else:
    print("\n⚠️ No specialties extracted. Check if website layout changed.")

driver.quit()
```

```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# ✅ Path to your local ChromeDriver
service = Service('C:\\Users\\chromedriver-win64\\chromedriver.exe')
driver = webdriver.Chrome(service=service)

# ✅ Open the website
driver.get('https://www.doccafe.com/crna-jobs')
driver.maximize_window()
wait = WebDriverWait(driver, 20)

# ✅ Click the Specialty dropdown
try:
    dropdown = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[contains(text(),'Specialty')]")))
    dropdown.click()
    print("✅ Clicked Specialty dropdown.")
except Exception as e:
    print("❌ Failed to click dropdown:", e)
    driver.quit()
    exit()

# ✅ Pause to let the menu render fully
time.sleep(4)

# ✅ Try to locate dropdown options dynamically
try:
    options = driver.find_elements(By.XPATH, "//div[@role='listbox']//div[@role='option']")
    print(f"✅ Found {len(options)} specialty options.")

    if not options:
        print("❌ No options found. Possible DOM issue.")

    specialties = [opt.text.strip() for opt in options if opt.text.strip()]

    # ✅ Print and save to CSV
    print("🩺 Extracted Specialties:")
    for idx, spec in enumerate(specialties, 1):
        print(f"{idx}. {spec}")

    with open('specialties.csv', 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(['Specialty'])
        for spec in specialties:
            writer.writerow([spec])

    print("\n✅ Saved to 'specialties.csv'")

except Exception as e:
    print("❌ Could not extract specialty options. Error:", e)

driver.quit()
```

```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# ✅ ChromeDriver path
service = Service('C:\\Users\\chromedriver-win64\\chromedriver.exe')
driver = webdriver.Chrome(service=service)
driver.maximize_window()
wait = WebDriverWait(driver, 20)

# ✅ List of job roles with URL and output file
job_roles = {
    "CRNA": "https://www.doccafe.com/crna-jobs",
    "Physician": "https://www.doccafe.com/physician-jobs",
    "Midwife": "https://www.doccafe.com/midwife-jobs",
    "Nurse Practitioner": "https://www.doccafe.com/nurse-practitioner-jobs",
    "Physician Assistant": "https://www.doccafe.com/physician-assistant-jobs"
}

for role, url in job_roles.items():
    try:
        print(f"\n🔍 Processing Occupation: {role}")
        driver.get(url)
        time.sleep(3)

        # ✅ Click the Specialty dropdown
        try:
            dropdown = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[contains(text(),'Specialty')]")))
            dropdown.click()
            print("✅ Clicked Specialty dropdown.")
        except Exception as e:
            print(f"❌ Failed to click dropdown for {role}: {e}")
            continue

        # ✅ Wait and collect all Specialty options
        time.sleep(4)  # wait for animation
        options = driver.find_elements(By.XPATH, "//div[@role='listbox']//div[@role='option']")
        print(f"✅ Found {len(options)} specialties under {role}")

        specialties = [opt.text.strip() for opt in options if opt.text.strip()]
        if not specialties:
            print("⚠️ No specialties extracted.")
            continue

        # ✅ Print the specialties
        for spec in specialties:
            print(f" ➤ {spec}")

        # ✅ Save to CSV
        filename = f"{role.lower().replace(' ', '_')}_specialties.csv"
        with open(filename, 'w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow(['Specialty'])
            for spec in specialties:
                writer.writerow([spec])

        print(f"💾 Saved to: {filename}")

    except Exception as e:
        print(f"❌ Error while processing {role}: {e}")

# ✅ Close browser after loop ends
driver.quit()
print("\n🎉 All occupations processed successfully.")
```
